```python
raw_data = spreadsheets.get_worksheet_by_id('145387737').get_all_values()
data_after_skip = raw_data[2:]
```

```python
df = pd.DataFrame(data_after_skip)
```

```python
df[:5]
```

```python
blank_col =[]
for i in range(6,74):
  if df.iloc[0,i] =="":
    blank_col.append(i)
print(blank_col)
```

```python
df.drop(df.columns[blank_col], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df.shape
```

```python
blank_rows =[3]
for i in range(4,len(df)):
  if df.iloc[i,1] == "" or "total" in df.iloc[i,1].lower() :
    blank_rows.append(i)
print(blank_rows)
```

```python
df2 = df.drop(index = blank_rows)
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2.head(5)
```

```python
item_df = df2.iloc[3:,:6].reset_index(drop=True)
```

```python

item_df.columns =  ["item"] + item_df.iloc[0, 1:].tolist()
```

```python
item_df = item_df[1:].rename_axis('item_id').reset_index()
```

```python
item_df.head(5)
```

```python
process_df = df2.iloc[:4,7:].T.reset_index(drop=True)
```

```python
process_df.columns = ["step1","step2","step3","step4"]
```

```python
process_df = process_df.rename_axis('process_id').reset_index()
```

```python
process_df.head(5)
```

```python
df2.shape
```

```python

def gen_newrow(df2,i,j):
  newrow = [df2.iloc[i,0],df2.iloc[i,1],df2.iloc[i,2],df2.iloc[i,3],df2.iloc[i,4],df2.iloc[i,5],df2.iloc[0,j],df2.iloc[1,j],df2.iloc[2,j],df2.iloc[3,j],df2.iloc[i,j]]
  return newrow
```

```python
col_names = gen_newrow(df2,3,7)
col_names[-1],col_names[-2],col_names[-3],col_names[-4],col_names[-5]= 'Values',"Process4","Process3","Process2","Process1"
print(col_names)
```

```python
templist = []
for i in range (4,df2.shape[0]):
  for j in range (6,df2.shape[1]):
    templist.append(gen_newrow(df2,i,j))
```

```python
len(templist)
```

```python
final_df = pd.DataFrame(templist[1:], columns=col_names)
```

```python
final_df.info()
```

```python
# unique_values = final_df['Values'].unique()
# print(unique_values)
```

```python
final_df['Values'] = final_df['Values'].str.replace(' ', '', regex=False)
```

```python
final_df['Values'] = final_df['Values'].str.replace(',', '', regex=False)
```

```python
final_df['Values'] = final_df['Values'].replace('', 0)
```

```python
final_df['Values'] = final_df['Values'].fillna(0)
```

```python
final_df['Values'] = final_df['Values'].replace('-', 0)
```

```python
final_df['Values'] = final_df['Values'].astype(float)
```
